## Q1. Line Plot (EDA Basics)

**Topic:** Simple line plots, `plt.plot`, labels

### Data (time-series style)

Pythondays = [1, 2, 3, 4, 5, 6, 7]daily_users = [120, 150, 180, 220, 260, 300, 360]

### Question

You are tracking **daily active users** for a new app.

1. Plot `daily_users` vs `days`
    
2. Add:
    
    * x-label: `"Day"`
        
    * y-label: `"Active Users"`
        
    * title: `"Daily Active Users Growth"`

## Q8. Error Bars (Uncertainty Visualization)

**Topic:** Error bars

### Data

Pythondays = [1, 2, 3, 4, 5]mean_sales = [200, 220, 250, 270, 300]std_dev = [20, 25, 15, 30, 20]

### Question

1. Plot mean sales with **error bars** using standard deviation.
    
2. Why is showing uncertainty **critical in data science reports**?

```python
days = [1, 2, 3, 4, 5]
mean_sales = [200, 220, 250, 270, 300]
std_dev = [20, 25, 15, 30, 20]

plt.errorbar(days, mean_sales, yerr=std_dev)
```

## Q10. Full EDA Visualization (Mini Real-World Task)

**Topic:** Combined concepts (lines, scatter, hist, subplots)

### Data

Pythondays = [1, 2, 3, 4, 5, 6, 7]users = [120, 150, 180, 220, 260, 300, 360]session_time = [15, 18, 20, 22, 25, 28, 30]

### Question

Create a **single figure** with:

1. Line plot: users vs days
    
2. Scatter plot: users vs session_time
    
3. Histogram: session_time
    

Then answer:

* What story does the data tell?
    
* What would be your **next step as a data scientist**?

```python
days = [1, 2, 3, 4, 5, 6, 7]
users = [120, 150, 180, 220, 260, 300, 360]
session_time = [15, 18, 20, 22, 25, 28, 30]

plt.figure(figsize=(18,4))

plt.subplot(1, 3, 1)
plt.plot(days, users)
plt.title("User Growth")

plt.subplot(1, 3, 2)
plt.scatter(users, session_time)

plt.subplot(1, 3, 3)
plt.hist(session_time, bins=5, edgecolor='black')
```

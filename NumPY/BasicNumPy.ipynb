```python
import numpy as np
```

## 1. Create a Numpy array of 100 random numbers and find mean, median and variance.

```python
arr = (np.random.randint(100, size = 100))

print(f"Mean: {np.mean(arr)}")
print(f"Median: {np.median(arr)}")
print(f"Variance: {np.var(arr)}")
```

## 3. Given a 2D array of sales data (rows=days, cols=products), find total sales per product.

```python
arr = np.array([np.arange(1, 10), np.random.randint(1000, size=9)])
print(arr)
np.sum(arr[1], axis = 0)
```

## 4. Create a 5x5 matrix of random integers and replace all values < 10 with 0

```python
arr = (np.random.randint(100, size=25)).reshape((5, 5))
arr[arr < 10] = 0
arr
```

## 5. Create an array of temperatures in Celsius and convert to Farenheit using vectorization.

```python
cel = np.array([34 ,54, 23, 54, 32, 6, 43, 76])

far = (5/9)*(cel+32)

print(f"Celcius: {cel}")
print(f"Farenheit: {far}")
```

## 6. Create an 1D array and  reverse it without using a loop.

```python
arr = np.random.randint(100, size=10)
print(f"{arr} \n{arr[::-1]}")
```

## 7. Generate 1000 random ages(18-60) and how many are above 40

```python
age = np.random.randint(18, 61, size=1000)
print(age)
print(np.sum(age > 40))
```

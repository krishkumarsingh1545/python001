## 1. Convert a week's Celcius temperature into Farenheit

```python
celWeek = (np.random.uniform(15, 40, 7))
farhenWeek = 9/5 * celWeek + 32
print(f"Celsius: {celWeek}")
print(f"Farenheit: {farhenWeek}")
```
